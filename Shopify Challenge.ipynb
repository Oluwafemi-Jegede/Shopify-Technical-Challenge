### 1(a) Think about what could be going wrong with our calculation. Think about a better way to evaluate this data.

**Looking at the first few items on the dataset, the following could be observed.**

- There is an issues with the `order_amount` and `total_items` recorded for a couple of stores(42 and 78) and this affecting the AOV calulation


- The problematic transaction for shop_id 42 comes from user_id 607 which are all credit card transactions. The items in question have same `order_amount` and `total_items`,  this is most likely a system glitch.


- As regards the shop-78, the error comes only from the `order_amount`, which are absurdly high 


**A better way to calculate the AOV be to take one of the following approach**

- Isolate `stores 78 and 42`, perform a root cause analysis on the issues in these stores (determine the actual price per unit for each stores and fix issues - fill unit prices with average price across board).


- Calculate AOV, excluding stores 78 and 42

The first approach will be taken 

```python
#adding the price per unit column

df['price_per_unit'] = df['order_amount']/df['total_items']

#re-arrange columns 
col = ['order_id', 'shop_id', 'user_id', 'order_amount', 'total_items','price_per_unit',
       'payment_method', 'created_at']

df = df[col]
```

- **Deep diving into both shops and  checking the price per unit**

```python
df[df['shop_id'] == 42]
```

- Even with the system error the price per unit is consistent at `$352`

The `order_amount` can be corrected to the right value `$704` and the `total_items` `which appears to be 2 instead of 2000`

```python
df['order_amount'] == 704000
```

```python
# correcting the order amount and total items
df.loc[df['order_amount'] == 704000,'order_amount'] = 704
df.loc[df['total_items'] == 2000,'total_items'] = 2
```

```python
df[df['shop_id'] == 42]
```

- Issues with shop 42 is now fixed

Next is shop 78

```python
df[df['shop_id'] == 78]
```

Appears there is little one can do about shop 78, without contacting the store or filling the `price_per_unit` is  with the `average price across other stores`

```python
# calculating the median of the prices across board and fill unit price for shop 78
df.loc[df['shop_id'] == 78,'price_per_unit'] = df[(df['shop_id'] != 78)]['price_per_unit'].mean()
df[df['shop_id'] == 78]
```

```python
# correcting the order_amount

df.loc[df['shop_id'] == 78,
       'order_amount'] = df[(df['shop_id'] == 78)]['total_items'] * df[(df['shop_id'] != 78)]['price_per_unit'].mean()


df[df['shop_id'] == 78]
```

Now shop-78 has been fixed

```python
new_AOV = df['order_amount'].mean()
new_AOV
```

**The new AOV figure is `~$304`**

```python
# calculating the AOV per store

df.groupby('shop_id')['order_amount'].mean().sort_values()
```

### 2(a) How many orders were shipped by Speedy Express in total?


- SQL code

`SELECT count(*) as num_of_orders
FROM Orders o
LEFT JOIN Shippers s on 
o.ShipperID = s.ShipperID
WHERE ShipperName = 'Speedy Express'
;`

**`54` orders were shipped by Speedy Express**

### 2(b) What is the last name of the employee with the most orders?

- SQL Code

`SELECT e.LastName
FROM
(
SELECT EmployeeID, count(*) as num_of_orders
FROM Orders
GROUP BY EmployeeID
ORDER BY num_of_orders DESC
LIMIT 1) as emp_o
LEFT JOIN Employees e 
on emp_o.EmployeeID = e.EmployeeID
;`


**The last name of the employee with the mist order is `Peacock` with `40 orders`**

### 2(c) What product was ordered the most by customers in Germany?


-- The problem will be approached by using CTEs; 1- ordersDetails which includes products, and 2- For Germany Orders


-- CTE 1 - order_details

`
WITH order_details AS
(
select *
from OrderDetails od
left Join Products p on 
od.ProductID = p.ProductID
)`
,


-- CTE 2 - Germany Order

`
ger_orders AS
(SELECT * 
FROM Orders o
Left JOIN Customers c on
o.CustomerID = c.CustomerID
where Country = 'Germany')
`

-- Bring it together and outputing results

`
select ProductName
from (
select os.ProductName, count(*) as num_of_orders
from ger_orders g 
left join order_details os
on g.OrderID = os.OrderID
group by os.ProductName
order by num_of_orders DESC
limit 1) as p_list;
`


**The most ordered product in Germany is `Gorgonzola Telino`**
